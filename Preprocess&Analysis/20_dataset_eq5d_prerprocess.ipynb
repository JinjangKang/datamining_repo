```python
data['HE_BMI'].value_counts()
# x축과 y축을 만듭니다.
x = data['HE_BMI'].value_counts().index
y = data['HE_BMI'].value_counts()

# 막대 그래프를 만듭니다.
plt.scatter(x, y)

# 그래프를 표시합니다.
plt.show()
```

```python
median_bmi = data['HE_BMI'].median()
data['HE_BMI'].fillna(median_bmi, inplace=True)
data['HE_BMI'].value_counts()
```
