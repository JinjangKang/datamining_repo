```python
data['BD1_11'].replace({8: 1, 9: 1}, inplace=True)
data['BD2_1'].replace({8: 0, 9: 0}, inplace=True)

median_sleep = data['BP16_1'].median()
data['BP16_1'].replace({88: median_sleep, 99: median_sleep}, inplace=True)

data['BS3_1'].replace({8: 0, 1: 3, 2:2, 3:1}, inplace=True)
data['BS3_2'].replace({888: 0, 999: 0}, inplace=True)

data['BE8_1'].replace({88: 0, 99: 0}, inplace=True)
data['BE3_31'].replace({88: 1, 99: 1}, inplace=True)

median_walk = data['BE3_32'].median()
data['BE3_32'].replace({88: median_walk, 99: median_walk}, inplace=True)

data['BE5_1'].replace({8: 1, 9: 1}, inplace=True)

data['HE_BMI'] = data['HE_BMI'].round(1)
```

```python
for i in data.columns :
    print(data[i].value_counts())
```

```python
data.to_csv('../datasets/20_dataset_2.csv', index=True)
```
